```python
import pandas as pd
import numpy as np

POPULATIONPATH = './data/pref_population_2015.csv'
INFLOWPATH = './data/inflow_hour20180101_20210228.npy'
OUTFLOWPATH = './data/outflow_hour20180101_20210228.npy'
OD_DAYS = [date.strftime('%Y-%m-%d %H:%M:%S') for date in pd.date_range(start='2018-01-01 00:00:00', end='2021-02-28 23:59:59', freq='1H')]

inflow, outflow = np.load(INFLOWPATH), np.load(OUTFLOWPATH)
print(inflow.shape, outflow.shape)
```

```python
START_DATE, END_DATE = '2019-09-22 09:00:00', '2019-10-27 08:00:00'
TARGET_AREA = ['東京都', '千葉県','神奈川県','埼玉県']
TARGET_AREA_EN = ['Tokyo', 'Chiba', 'Kanagawa', 'Saitama']
if len(TARGET_AREA) == 0:
    pref_population = pd.read_csv(POPULATIONPATH)
    TARGET_AREA = pref_population['ken'].values
    POPULATION = pref_population['population'].values
    TARGET_AREA_ID = np.arange(POPULATION).tolist()
    print(POPULATION)
    print(TARGET_AREA_ID)
else:
    pref_population = pd.read_csv(POPULATIONPATH)
    pref_population['gid'] -= 1
    pref_population = pref_population[pref_population['ken'].isin(TARGET_AREA)]
    pref_population = pref_population[['ken', 'gid', 'population']]
    pref_population = pref_population.set_index('ken')
    pref_population = pref_population.reindex(index=TARGET_AREA)
    POPULATION = pref_population['population'].values.squeeze()
    TARGET_AREA_ID = pref_population['gid'].values.squeeze().tolist()
    print(POPULATION)
    print(TARGET_AREA_ID)
    
hagibis_inflow = inflow[OD_DAYS.index(START_DATE):OD_DAYS.index(END_DATE)+1,:]
hagibis_inflow = hagibis_inflow[:, TARGET_AREA_ID]
hagibis_outflow = outflow[OD_DAYS.index(START_DATE):OD_DAYS.index(END_DATE)+1,:]
hagibis_outflow = hagibis_outflow[:, TARGET_AREA_ID]

print(hagibis_inflow.shape, hagibis_outflow.shape)
```

```python
hagibis_twitter = pd.read_csv('./data/Hagibis_twitter_count_by_prefecture.csv')
hagibis_twitter.columns = ['time', '神奈川県', '栃木県', '千葉県', '東京都', '埼玉県', '茨城県', '群馬県']
hagibis_twitter
```

```python
hagibis_twitter['time'] = pd.to_datetime(hagibis_twitter['time'])
hagibis_twitter['time'] = hagibis_twitter['time'] + pd.Timedelta(hours=9)
hagibis_twitter = hagibis_twitter.sort_values(by='time')
hagibis_twitter['time'] = hagibis_twitter.time.dt.strftime('%Y-%m-%d %H:%M:%S')
hagibis_twitter
```

```python
hagibis_timestamps = [date.strftime('%Y-%m-%d %H:%M:%S') for date in pd.date_range(START_DATE, END_DATE, freq='1H')]
len(hagibis_timestamps), hagibis_timestamps[0], hagibis_timestamps[-1]
```

```python
hagibis_twitter = pd.merge(hagibis_twitter, pd.DataFrame({'time': hagibis_timestamps}), how='right')
hagibis_twitter = hagibis_twitter.fillna(0)
hagibis_twitter = hagibis_twitter[TARGET_AREA]
hagibis_twitter
```

```python
hagibis_twitter = hagibis_twitter.values
hagibis_twitter.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot(pref_name):
    pref_index = TARGET_AREA.index(pref_name)
    pref_name_en = TARGET_AREA_EN[pref_index]
    hagibis_inflow_pref = hagibis_inflow[:, pref_index]
    hagibis_inflow_pref_max = np.max(hagibis_inflow_pref)
    hagibis_outflow_pref = hagibis_outflow[:, pref_index]
    hagibis_outflow_pref_max = np.max(hagibis_outflow_pref)
    hagibis_twitter_pref = hagibis_twitter[:, pref_index]
    hagibis_twitter_pref_max = np.max(hagibis_twitter_pref)
    
    plt.figure(figsize =(25, 8)) 
    plt.plot(hagibis_inflow_pref/hagibis_inflow_pref_max)
    plt.plot(hagibis_outflow_pref/hagibis_outflow_pref_max)
    plt.plot(hagibis_twitter_pref/hagibis_twitter_pref_max)
    plt.legend(loc="best", labels=[f'{pref_name_en} Hourly Inflow Mobility, MaxValue={hagibis_inflow_pref_max}', 
                                   f'{pref_name_en} Hourly Outflow Mobility, MaxValue={hagibis_outflow_pref_max}', 
                                   f'{pref_name_en} Hourly Disaster-Related Twitter, MaxValue={hagibis_twitter_pref_max}'])
    plt.title(f"Normalized Inflow, Outflow, Twitter at {pref_name_en}", fontsize=16, y=1)
    plt.xticks(np.arange(0, len(hagibis_timestamps), 12), hagibis_timestamps[::12], rotation=90)
    plt.xlabel('Time', fontsize=16)
    plt.ylabel('Normalized Number', fontsize=16)

    plt.show()
```

```python
for pref in TARGET_AREA:
    plot(pref)
```
