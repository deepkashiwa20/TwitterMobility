```python
import pandas as pd
```

```python
good_twitter = pd.read_csv('./data/Twitter_Hagibis2019.csv', index_col=0)
good_twitter['CreateTime'] = pd.to_datetime(good_twitter['CreateTime']) 
good_twitter['Hour'] = good_twitter['CreateTime'] + pd.Timedelta(hours=9)
good_twitter['Hour'] = good_twitter['Hour'].dt.floor('H')
good_twitter
```

```python
START_DATE, END_DATE = '2019-09-22 09:00:00', '2019-10-27 08:00:00'
timeslices = [date.strftime('%Y-%m-%d %H:%M:%S') for date in pd.date_range(START_DATE, END_DATE, freq='1H')]
len(timeslices), timeslices[0], timeslices[-1]
```

```python
hour_twitter = good_twitter.groupby(['Hour'], as_index=False).size()
hour_twitter = hour_twitter.sort_values(by='Hour')
hour_twitter['Hour'] = hour_twitter.Hour.dt.strftime('%Y-%m-%d %H:%M:%S')
hour_twitter = pd.merge(hour_twitter, pd.DataFrame({'Hour': timeslices}), how='right')
hour_twitter = hour_twitter.fillna(0)
hour_twitter
```

```python
Twitter = hour_twitter['size'].values
Twitter.shape
```

```python
import scipy.sparse as ss
import numpy as np

POPULATIONPATH = './data/pref_population_2015.csv'
ODPATH = './data/od_hour20180101_20210228.npz'
OD_DAYS = [date.strftime('%Y-%m-%d %H:%M:%S') for date in pd.date_range(start='2018-01-01 00:00:00', end='2021-02-28 23:59:59', freq='1H')]
print(len(OD_DAYS))
TARGET_AREA = ['東京都', '千葉県','神奈川県','埼玉県']
TARGET_AREA_EN = ['Tokyo', 'Chiba', 'Kanagawa', 'Saitama']

if len(TARGET_AREA) == 0:
    pref_population = pd.read_csv(POPULATIONPATH)
    TARGET_AREA = pref_population['ken'].values
    POPULATION = pref_population['population'].values
    TARGET_AREA_ID = np.arange(POPULATION).tolist()
    print(POPULATION)
    print(TARGET_AREA_ID)
else:
    pref_population = pd.read_csv(POPULATIONPATH)
    pref_population['gid'] -= 1
    pref_population = pref_population[pref_population['ken'].isin(TARGET_AREA)]
    pref_population = pref_population[['ken', 'gid', 'population']]
    pref_population = pref_population.set_index('ken')
    pref_population = pref_population.reindex(index=TARGET_AREA)
    POPULATION = pref_population['population'].values.squeeze()
    TARGET_AREA_ID = pref_population['gid'].values.squeeze().tolist()
    print(POPULATION)
    print(TARGET_AREA_ID)
    
trans = ss.load_npz(ODPATH)
trans = np.array(trans.todense()).reshape((-1, 47, 47))
trans = trans[OD_DAYS.index(START_DATE):OD_DAYS.index(END_DATE)+1,:,:]
trans = trans[:, TARGET_AREA_ID, :][:, :, TARGET_AREA_ID]
print(trans.shape)
```

```python
import matplotlib.pyplot as plt
def plot(OD, OD_name, Twitter):
    OD_max = np.max(OD)
    Twitter_max = np.max(Twitter)
    plt.figure(figsize =(25, 8)) 
    plt.plot(Twitter/Twitter_max)
    plt.plot(OD/OD_max)
    plt.legend(loc="best", labels=[f'Hourly Disaster-Related Twitter (all prefectures), MaxValue={Twitter_max}', f'Hour {OD_name}, MaxValue={OD_max}'])
    plt.title(f'Normalized Twitter and {OD_name}', fontsize=16, y=1)
    plt.xticks(np.arange(0, len(timeslices), 12), timeslices[::12], rotation=90)
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.show()
```

```python
OD_Chiba2Tokyo = trans[:, 1, 0]
plot(OD_Chiba2Tokyo, 'OD_Chiba2Tokyo', Twitter)
```

```python
OD_Kanagawa2Tokyo = trans[:, 2, 0]
plot(OD_Kanagawa2Tokyo, 'OD_Kanagawa2Tokyo', Twitter)
```
